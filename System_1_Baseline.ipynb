```python
import pandas as pd
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("1. Downloading GoEmotions dataset...")
dataset = load_dataset("google-research-datasets/go_emotions", split="train")
df = pd.DataFrame(dataset)

print("2. Cleaning the data...")
df['primary_emotion'] = df['labels'].apply(lambda x: x[0] if len(x) > 0 else 27)
df_sample = df.sample(n=20000, random_state=42)

print("3. Splitting data into Training and Testing sets...")
X_train, X_test, y_train, y_test = train_test_split(
    df_sample['text'],
    df_sample['primary_emotion'],
    test_size=0.2,
    random_state=42
)

print("4. Vectorizing text (turning English words into mathematical frequencies)...")
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

print("5. Training the System 1 Baseline Model (Logistic Regression)...")
model = LogisticRegression(max_iter=1000)
model.fit(X_train_vectorized, y_train)

print("6. Testing the model...")
predictions = model.predict(X_test_vectorized)
accuracy = accuracy_score(y_test, predictions)

print(f"\n======================================")
print(f"Baseline Model Accuracy: {accuracy * 100:.2f}%")
print(f"======================================")
print("Conclusion: Distinguishing between 28 fine-grained emotions using only word frequencies is highly inaccurate.")
```
